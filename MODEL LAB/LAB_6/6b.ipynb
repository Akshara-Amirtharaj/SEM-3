```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv(r"C:\Users\aksha\OneDrive - SSN Trust\Desktop\AKSHARA\COLLEGE\SEM 3\EDDA\MODEL LAB\LAB_6\data.csv")
dataset
```

```python
dataset.drop("Unnamed: 32",axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
dataset["diagnosis"]=encoder.fit_transform(dataset["diagnosis"])
dataset
```

```python
x=dataset.drop("diagnosis",axis=1)
y=dataset["diagnosis"]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca=PCA()
x_new=pca.fit(x_scaled)
```

```python
cumulative_sum=np.cumsum(pca.explained_variance_ratio_)
```

```python
import matplotlib.pyplot as plt
plt.plot(cumulative_sum)
```

```python
pca=PCA(n_components=2)
x_new=pca.fit_transform(x_scaled)
x_newdf=pd.DataFrame(data=x_new,columns=["pc1","pc2"])
x_newdf["diagnosis"]=dataset["diagnosis"]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot with Seaborn to use 'hue'
sns.scatterplot(x='pc1', y='pc2', hue='diagnosis', data=x_newdf)

# Show plot
plt.show()
```
