```python
# Fill missing values in 'Salary' column with the mean salary
dataset["Salary"].fillna(dataset["Salary"].mean(), inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Create an instance of OneHotEncoder without sparse output
encoder = OneHotEncoder(sparse_output=False)

# Fit and transform the selected columns
encoded = encoder.fit_transform(dataset[["League", "Division", "NewLeague"]])

# Convert the encoded array into a DataFrame
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(["League", "Division", "NewLeague"]))

# Display the DataFrame
encoded_df
```

```python
dataset.drop(["League", "Division", "NewLeague"],axis=1,inplace=True)
dataset
```

```python
dataset=pd.concat([dataset,encoded_df],axis=1)
dataset
```

```python
from scipy.stats import zscore
import numpy as np
z_scores=np.abs(zscore(dataset))
outliers=np.where(z_scores>3)
```

```python
outliers
```

```python
x=dataset.drop("Salary",axis=1)
y=dataset["Salary"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.linear_model import Lasso,Ridge,LinearRegression
model=Lasso(alpha=0.1)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model=Ridge(alpha=0.1)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```
