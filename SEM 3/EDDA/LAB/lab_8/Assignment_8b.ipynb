```python
import pandas as pd
import numpy as np
from scipy.stats import zscore
```

```python
dataset=pd.read_csv('Hitters.csv')
dataset
```

```python
dataset.describe
```

```python
dataset.isnull().sum()
```

```python
dataset["Salary"]=dataset["Salary"].fillna(dataset["Salary"].mean())
dataset
```

```python
dataset.drop("Unnamed: 0",axis=1,inplace=True)
dataset
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder(sparse=False)
encoded=encoder.fit_transform(dataset[["League","Division","NewLeague"]])
encoded_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out(["League","Division","NewLeague"]))
encoded_df
```

```python
dataset.drop(["League","Division","NewLeague"],axis=1,inplace=True)
dataset
```

```python
dataset=pd.concat([dataset,encoded_df],axis=1)
dataset
```

```python
z_scores = np.abs(zscore(dataset))
outliers = np.where(z_scores > 3)
outliers
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score,accuracy_score
x=dataset.drop("Salary",axis=1)
y=dataset["Salary"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import Lasso
lasso_model=Lasso(alpha=0.1)
lasso_model.fit(x_train,y_train)
lasso_model.score(x_test,y_test)
```

```python
from sklearn.linear_model import Ridge
ridge_model=Ridge(alpha=0.1)
ridge_model.fit(x_train,y_train)
ridge_model.score(x_test,y_test)
```

```python
from sklearn.linear_model import LinearRegression
linear_model=LinearRegression()
linear_model.fit(x_train,y_train)
linear_model.score(x_test,y_test)
```
