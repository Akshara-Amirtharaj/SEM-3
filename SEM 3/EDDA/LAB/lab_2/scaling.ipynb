```python
import numpy as np
import pandas as pd
```

```python
#2
from sklearn.preprocessing import RobustScaler
x=np.array([[1,-2,2],[-2,1,3],[4,1,2]])
scaler=RobustScaler()
robust_scaled=scaler.fit_transform(x)
robust_scaled=pd.DataFrame(robust_scaled)
robust_scaled
```

```python
#3
from sklearn.preprocessing import StandardScaler
y=np.array([[1,-1,2],[2,0,0],[0,1,-1]])
scaler=StandardScaler()
standard_scaled=scaler.fit_transform(y)
standard_scaled=pd.DataFrame(standard_scaled)
standard_scaled
```

```python
#4
from sklearn.preprocessing import MaxAbsScaler
z=np.array([[1,-1,2],[2,0,0],[0,1,-1]])
scaler=MaxAbsScaler()
max_scaled=scaler.fit_transform(z)
max_scaled=pd.DataFrame(max_scaled)
max_scaled
```

```python
#5
from sklearn.preprocessing import OneHotEncoder
data=[['Male', 1], ['Female', 3], ['Female', 2]]
df=pd.DataFrame(data)
categorical_columns=df.select_dtypes(include=['object']).columns.tolist()
encoder=OneHotEncoder(sparse=False)
one_hot_encoded_data=encoder.fit_transform(df[categorical_columns])
one_hot_encoded_data
```
